```python
# Import libraries
import os

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Read News and DJIA Data
df_news = pd.read_csv(os.path.join('..', 'data', 'Combined_News_DJIA.csv'))
df_stock = pd.read_csv(os.path.join('..', 'data', 'upload_DJIA_table.csv'))
```

```python
df_news.drop('Label', axis=1, inplace=True)
```

```python
# Adjust display settings

pd.set_option('display.max_colwidth', None)
pd.set_option('display.expand_frame_repr', False)

# Get 5 random samples with reproducibility
df_news.sample(n=5, random_state=42)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import torch

def FinBERT_sentiment_score(headings):
    """
    Compute the mean sentiment score using pretrained FinBERT on -1 to 1 scale. 
    -1 being negative and 1 being positive.
    
    Parameters:
        headings (list of str): A list of text headings.
        
    Returns:
        float: The mean sentiment score of all input headings.
    """

    # Specify the device (GPU if available)
    device = 0 if torch.cuda.is_available() else -1
    
    # Load tokenizer and model
    tokenizer = AutoTokenizer.from_pretrained('ProsusAI/finbert')
    finbert = AutoModelForSequenceClassification.from_pretrained('ProsusAI/finbert')
    
    # Initialize sentiment analysis pipeline with FinBERT
    nlp = pipeline("sentiment-analysis", model=finbert, tokenizer=tokenizer, device=device)
    
    scores = []
    for heading in headings:
        result = nlp(heading)
        if result[0]['label'] == "positive":
            scores.append(result[0]['score'])
        elif result[0]['label'] == "neutral":
            scores.append(0)
        else:
            scores.append(-result[0]['score'])
    
    mean_score = sum(scores) / len(scores) if scores else 0
    
    return mean_score

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
    
def VADER_sentiment_score(headings):
    """
    compute sentiment score using pretrained VADER on -1 to 1 scale. -1 being negative and 1 being positive
    """

    
    analyzer = SentimentIntensityAnalyzer()
    scores = []
    
    for heading in headings:
        result = analyzer.polarity_scores(heading)
        if result['pos'] == max(result['neg'], result['neu'], result['pos']):
            scores.append(result['pos'])
        elif result['neg'] == max(result['neg'], result['neu'], result['pos']):
            scores.append(-result['neg'])
        else:
            scores.append(0)
    
    mean_score = sum(scores) / len(scores) if scores else 0
    
    return mean_score


news_df = df_news


BERT_sentiment = []
VADER_sentiment = []


for i in range(len(news_df)):
    print(f'Processing {i+1}/{len(news_df)}')
    news_list = news_df.iloc[i, 1:].tolist()
    news_list = [i for i in news_list if i != '0']
    score_BERT = FinBERT_sentiment_score(news_list)
    score_VADER = VADER_sentiment_score(news_list)
    BERT_sentiment.append(score_BERT)
    VADER_sentiment.append(score_VADER)


# print(news_df.iloc[129])

news_df['FinBERT score'] = BERT_sentiment
news_df['VADER score'] = VADER_sentiment

news_df.to_csv(os.path.join('..', 'data',"sentiment.csv"))
```
